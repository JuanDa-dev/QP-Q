```python
import pylatexenc as pl
from qiskit import QuantumCircuit

# Dibuja la matriz de la compuerta X 
qc3 = QuantumCircuit(1)
qc3.x(0)

simulator = Aer.get_backend('unitary_simulator')
compiled_circuit = transpile(qc3, simulator)
result = simulator.run(compiled_circuit).result()
unitary = result.get_unitary(compiled_circuit)

# Visualizar la matriz resultante
array_to_latex(unitary, prefix="\\text{Matriz resultante: }")
```
